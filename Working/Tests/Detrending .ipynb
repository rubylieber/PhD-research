```python
import xarray as xr
import scipy.signal as ss
import matplotlib.pyplot as plt 
```

```python
# Open HadISST 
sst_ds = xr.open_dataset('/g/data/w40/rl5183/hadisst.1870_2017_1.5x1.5.nc')
sst = sst_ds.sst.sel(lat=slice(-5,5)).sel(lon=slice(190,240)).sel(time=slice('1900-01', '2014-12')).mean(dim=['lat', 'lon'])
```

```python
sst.plot(figsize=(15,5))
```

```python
dt = ss.detrend(sst)
fig = plt.figure(figsize=(15,5))
plt.plot(dt)
```

```python
b = ss.bspline(sst, 2)
fig = plt.figure(figsize=(15,5))
plt.plot(b)
```
