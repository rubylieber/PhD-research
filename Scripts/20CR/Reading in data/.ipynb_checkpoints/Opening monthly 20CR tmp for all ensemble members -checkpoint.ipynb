```python
import xarray as xr 
import dask.diagnostics
import climtas.io
```

```python
# Generate member list
members = [str(x).zfill(2) for x in range(1,81)]
```

```python
for m in members:
    # Generate path
    path = f'/g/data/ua8/LE_models/20CRv3/member_monthly/TMPS/*/*_mem0{m}.nc'
    # Read in all years for that member 
    ds = xr.open_mfdataset(path, parallel=True)
    # Chunk data 
    ds_chunk = ds.TMP.chunk(chunks={'time':-1, 'lat':110, 'lon':110})
    # Save as netcdf
    ds_chunk.to_netcdf(f'/g/data/w40/rl5183/20CR/temp/members/monthly_tmp_mem0{m}.nc')
```

```python
# Remove member 70 
members = [*range(1,70),*range(71,81)]
```

```python
tmp_paths = [f"/g/data/w40/rl5183/20CR/temp/members/monthly_tmp_mem0{m:02d}.nc" for m in members]
```

```python
ds_tmp = xr.open_mfdataset(tmp_paths, combine='nested', concat_dim='member', chunks={'time':-1, 'lat':110, 'lon':110})
ds_tmp.coords['member'] = members
```

```python
# Save as netcdf
with dask.diagnostics.ProgressBar():
    climtas.io.to_netcdf_throttled(ds_tmp, '/g/data/w40/rl5183/20CR/temp/monthly_tmp_all_members.nc')
```
