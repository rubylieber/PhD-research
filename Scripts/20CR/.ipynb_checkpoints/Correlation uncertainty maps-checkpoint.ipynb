```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
import xesmf
import cf_xarray
import cartopy.crs as ccrs
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
```

```python
first_half = xr.open_dataarray('/g/data/w40/rl5183/20CR/temp/tmp_ensemble_std_corr_1901_1950.nc')
second_half = xr.open_dataarray('/g/data/w40/rl5183/20CR/temp/tmp_ensemble_std_corr_1950_2000.nc')
hdi_ds = xr.open_dataset('/home/563/rl5183/Downloads/HDI_1990_2015_1x1.nc')
hdi = hdi_ds.HDI.sel(time=2015)
```

```python
pop = xr.open_dataset('/g/data/w40/rl5183/gpw_v4_population_density_rev11_1_deg.nc')
pop_2015 = pop.sel(raster=4) # raster 4 is population density for 2015 (see csv file in zip for details of other rasters)
```

```python
# Regrid population data 
ds_bnds = pop.cf.add_bounds(['latitude','longitude'])
sample_bnds = hdi_ds.cf.add_bounds(['lat','lon'])
regrid = xesmf.Regridder(ds_bnds, sample_bnds, method='conservative_normed')
pop_2015_rg = regrid(pop_2015)
```

```python
pop_data = pop_2015_rg.to_array().sel(variable='Population Density, v4.11 (2000, 2005, 2010, 2015, 2020): 1 degree')
```

```python
# regrid 
first_half = regrid(first_half)
second_half = regrid(second_half)
```

```python
mask = hdi * first_half[0]
hdi = hdi.where(np.isfinite(mask))
first_half = first_half.where(np.isfinite(mask))
second_half = second_half.where(np.isfinite(mask))
```

```python
fig = plt.figure(figsize=(15,7.5))
axs = []

ax = fig.add_subplot(221, projection=ccrs.Robinson(180))
C = first_half.sel(month=11).plot(ax=ax, add_colorbar=False, transform=ccrs.PlateCarree())
ax.coastlines()
ax.set_global()
ax.set_title('NDJ')
# Gridlines
gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=2, color='gray', alpha=0.5, linestyle='--')
gl.xlabels_top = False
gl.ylabels_right = False
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER
axs.append(ax)

ax = fig.add_subplot(222, projection=ccrs.Robinson(180))
C = first_half.sel(month=2).plot(ax=ax, add_colorbar=False, transform=ccrs.PlateCarree())
ax.coastlines()
ax.set_global()
ax.set_title('FMA')
# Gridlines
gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=2, color='gray', alpha=0.5, linestyle='--')
gl.xlabels_top = False
gl.ylabels_right = False
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER
axs.append(ax)

ax = fig.add_subplot(223, projection=ccrs.Robinson(180))
C = first_half.sel(month=5).plot(ax=ax, add_colorbar=False, transform=ccrs.PlateCarree())
ax.coastlines()
ax.set_global()
ax.set_title('MJJ')
# Gridlines
gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=2, color='gray', alpha=0.5, linestyle='--')
gl.xlabels_top = False
gl.ylabels_right = False
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER
axs.append(ax)

ax = fig.add_subplot(224, projection=ccrs.Robinson(180))
C = first_half.sel(month=8).plot(ax=ax, add_colorbar=False, transform=ccrs.PlateCarree())
ax.coastlines()
ax.set_global()
ax.set_title('ASO')
# Gridlines
gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=2, color='gray', alpha=0.5, linestyle='--')
gl.xlabels_top = False
gl.ylabels_right = False
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER
axs.append(ax)

# Colorbar
cbar = plt.colorbar(C, ax=axs, shrink=0.5, pad=0.05)
cbar.ax.set_ylabel('Pearsons r', fontsize=12)
```
