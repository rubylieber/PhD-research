```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
import xesmf
import cf_xarray
import cartopy.crs as ccrs
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
```

```python
hdi_ds = xr.open_dataset('/home/563/rl5183/Downloads/HDI_1990_2015_1x1.nc')
hdi = hdi_ds.HDI.sel(time=2015)
std_mean_tmp_first = xr.open_dataarray('/g/data/w40/rl5183/20CR/temp/tmp_ensemble_std_corr_1901_1950.nc')
std_mean_tmp_second = xr.open_dataarray('/g/data/w40/rl5183/20CR/temp/tmp_ensemble_std_corr_1950_2000.nc')
std_mean_pr_first = xr.open_dataarray('/g/data/w40/rl5183/20CR/precip/prate_ensemble_std_corr_1901_1950.nc')
std_mean_pr_second = xr.open_dataarray('/g/data/w40/rl5183/20CR/precip/prate_ensemble_std_corr_1950_2000.nc')
```

```python
pop = xr.open_dataset('/g/data/w40/rl5183/gpw_v4_population_density_rev11_1_deg.nc')
pop_2015 = pop.sel(raster=4) # raster 4 is population density for 2015 (see csv file in zip for details of other rasters)
```

```python
# Regrid population data 
ds_bnds = pop.cf.add_bounds(['latitude','longitude'])
sample_bnds = hdi_ds.cf.add_bounds(['lat','lon'])
regrid = xesmf.Regridder(ds_bnds, sample_bnds, method='conservative_normed')
pop_2015_rg = regrid(pop_2015)
```

```python
pop_data = pop_2015_rg.to_array().sel(variable='Population Density, v4.11 (2000, 2005, 2010, 2015, 2020): 1 degree')
pop_data.plot(vmax=100)
```

```python
# Use regridder on mean tmp std 
std_mean_tmp_first = regrid(std_mean_tmp_first)
std_mean_tmp_second = regrid(std_mean_tmp_second)
std_mean_pr_first = regrid(std_mean_pr_first)
std_mean_pr_second = regrid(std_mean_pr_second)
```

### Color code by latitude 

```python
# You need to get rid of the NaN. For this you first need to make sure all the data have the same land/sea mask.
# Create a mask that masks any NaN in the 3 fields, take only the first time for std_rx1_cor assuming the mask is the same for all times:
allmask = pop_data * hdi * std_mean_tmp_first[0]

# Apply to all fields. The following will keep the field where allmask is not NaN and put NaN otherwise.
hdi = hdi.where(np.isfinite(allmask))
pop_data = pop_data.where(np.isfinite(allmask))
std_mean_tmp_first = std_mean_tmp_first.where(np.isfinite(allmask))
std_mean_tmp_second = std_mean_tmp_second.where(np.isfinite(allmask))
std_mean_pr_first = std_mean_pr_first.where(np.isfinite(allmask))
std_mean_pr_second = std_mean_pr_second.where(np.isfinite(allmask))
```

```python
def ds_by_lat(ds):
    
    # Define the regions
    regions = {
    "Tropics":slice(-23.5,23.5),
    "Subtropics NH":slice(23.5,35.5),
    "Subtropics SH":slice(-35.5, -23.5),
    "Mid latitudes NH":slice(35.5, 66.5),
    "Mid latitudes SH":slice(-66.5, -35.5),
    }
    
    ds_by_lat = []
    for reg_name,reg_loc in regions.items():
        ds_sliced = ds.sel(lat=reg_loc).rename(reg_name)
        ds_by_lat.append(ds_sliced)
    ds_by_lat = xr.merge(ds_by_lat)
    return ds_by_lat
```

```python
hdi_by_lat = ds_by_lat(hdi)
pop_by_lat = ds_by_lat(pop_data)
tmp_first_by_lat = ds_by_lat(std_mean_tmp_first)
tmp_second_by_lat = ds_by_lat(std_mean_tmp_second)
pr_first_by_lat = ds_by_lat(std_mean_pr_first)
pr_second_by_lat = ds_by_lat(std_mean_pr_second)
```

### Weight dot size by population 

```python
# Define the regions
regions = {
"Tropics":slice(-23.5,23.5),
"Subtropics NH":slice(23.5,35.5),
"Subtropics SH":slice(-35.5, -23.5),
"Mid latitudes NH":slice(35.5, 66.5),
"Mid latitudes SH":slice(-66.5, -35.5),
}
```

```python
s={}
for reg_name, reg_loc in regions.items():
    s[reg_name] = (pop_by_lat[reg_name].stack({"z":["lat","lon"]}).dropna("z"))**2/10**4
```

## Calculate correlation coefficient

```python
hdi_1d = hdi.values.flatten()
hdi_1d = hdi_1d[np.isfinite(hdi_1d)]

tmp_first_1d = std_mean_tmp_first.sel(month=11).values.flatten()
tmp_first_1d = tmp_first_1d[np.isfinite(tmp_first_1d)]

tmp_second_1d = std_mean_tmp_second.sel(month=11).values.flatten()
tmp_second_1d = tmp_second_1d[np.isfinite(tmp_second_1d)]

pr_first_1d = std_mean_pr_first.sel(month=11).values.flatten()
pr_first_1d = pr_first_1d[np.isfinite(pr_first_1d)]

pr_second_1d = std_mean_pr_second.sel(month=11).values.flatten()
pr_second_1d = pr_second_1d[np.isfinite(pr_second_1d)]
```

```python
from scipy.stats import spearmanr
from scipy.stats import pearsonr

spear_tmp_first = spearmanr(hdi_1d, tmp_first_1d)
spear_tmp_second = spearmanr(hdi_1d, tmp_second_1d)

spear_pr_first = spearmanr(hdi_1d, pr_first_1d)
spear_pr_second = spearmanr(hdi_1d, pr_second_1d)

pear_tmp_first = pearsonr(hdi_1d, tmp_first_1d)
pear_tmp_second = pearsonr(hdi_1d, tmp_second_1d)

pear_pr_first = pearsonr(hdi_1d, pr_first_1d)
pear_pr_second = pearsonr(hdi_1d, pr_second_1d)
```
