```python
import xarray as xr
import matplotlib.pyplot as plt
import xesmf
import cf_xarray
import numpy as np
import cartopy.crs as ccrs
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
```

```python
obc_ds = xr.open_dataset('/g/data/w40/rl5183/obscount_20CRv3_rg.nc')
obc_data = obc_ds.obcount
```

```python
obc_data_2015 = obc_data.sel(time='2015-12')
```

```python
# HDI data 
hdi_ds = xr.open_dataset('/g/data/w40/rl5183/HDI_1990_2015_2x2.nc')
hdi = hdi_ds.HDI.sel(time='2015')
```

```python
# Uncertainty data 
std_ds = xr.open_dataset('/g/data/w40/rl5183/20CR/precip/prate_ensemble_std_corr_1901_2015.nc')
```

```python
# Regrid on a 2x2
ds_bnds = std_ds.cf.add_bounds(['latitude','longitude'])
sample_bnds = obc_ds.cf.add_bounds(['lon','lat'])
```

```python
regrid = xesmf.Regridder(ds_bnds, sample_bnds, method='conservative_normed')
```

```python
std_data = std_ds.__xarray_dataarray_variable__
```

```python
std = regrid(std_data)
```

```python
# Apply land mask to obs count data 
all_mask = hdi * obc_data_2015 * std[0]
```

```python
hdi_mask = hdi.where(np.isfinite(all_mask))
obc_mask = obc_data_2015.where(np.isfinite(all_mask))
std_mask = std.where(np.isfinite(all_mask))
```

```python
fig = plt.figure()
ax = plt.gca()
ax.scatter(obc_mask, std_mask.sel(month=11), s =1)
ax.set_xscale('log')
```

```python
# Define the regions
regions = {
    "Tropics":slice(-23.5,23.5),
    "Subtropics NH":slice(23.5,35.5),
    "Subtropics SH":slice(-35.5, -23.5),
    "Mid-latitudes NH":slice(35.5, 60),
    "Mid-latitudes SH":slice(-60, -35.5),
    "High-latitudes NH":slice(60,90),
}
```

```python
obc_mask_by_lat = []
for reg_name,reg_loc in regions.items():
    obc_mask_sliced = obc_mask.sel(lat=reg_loc).rename(reg_name)
    obc_mask_by_lat.append(obc_mask_sliced)
obc_mask_by_lat = xr.merge(obc_mask_by_lat)
```

```python
std_mask_by_lat = []
for reg_name,reg_loc in regions.items():
    std_mask_sliced = std_mask.sel(lat=reg_loc).rename(reg_name)
    std_mask_by_lat.append(std_mask_sliced)
std_mask_by_lat = xr.merge(std_mask_by_lat)
```

```python
fig = plt.figure(figsize=(10,5))

ax = fig.add_subplot(111)
for reg_name, reg_loc in regions.items():
    plt.scatter(obc_mask_by_lat[reg_name].stack({"z":["lat","lon"]}).dropna("z"),
                std_mask_by_lat[reg_name].sel(month=11).stack({"z":["lat","lon"]}).dropna("z"),
                s=2,label=reg_name, alpha=0.5)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.ylabel(''r'$\sigma$')
plt.gca().set_xscale('log')
plt.xlabel('Observations/Month')
lgnd = plt.legend(loc=0)
lgnd.legendHandles[0]._sizes = [50]
lgnd.legendHandles[1]._sizes = [50]
lgnd.legendHandles[2]._sizes = [50]
lgnd.legendHandles[3]._sizes = [50]
lgnd.legendHandles[4]._sizes = [50]
lgnd.legendHandles[5]._sizes = [50]
#plt.savefig('paper_fig_supp_3.jpg', bbox_inches='tight', dpi=300, quality=80, optimize=True, progressive=True)
```
